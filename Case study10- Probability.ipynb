```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for column in data.columns:
    data[column] = labelencoder.fit_transform(data[column])
```

```python
#The column "veil-type" is 0 and not contributing to the data so we can remove it
data = data.drop(["veil-type"],axis=1)
```

```python
#Setting X and y axis
y=data['class']
x=data.drop(['class'],axis=1)
```

```python
# splitting the data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

# Classification Methods

# 1. Decision Tree Classification

```python
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor()
dt_model.fit(x_train,y_train)
y_pred = dt_model.predict(x_test)
print('accuracy is:', dt_model.score(x_test,y_test)*100)
```

# 2. Logistic Regression Classification

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
print('accuracy is:', model.score(x_test,y_test)*100)
```

# 3.kNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print('accuracy is:', model.score(x_test,y_test)*100)
```

# 4.Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred=rf.predict(x_test)
from sklearn.metrics import accuracy_score
print('accuracy is:',accuracy_score(y_test,y_pred)*100)
```

# 5.SVM Classification

# 6. Naive Bayes Classification

```python
#fitting gaussian naive bayes to the training set
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy is:',accuracy_score(y_test,y_pred))

confusion_matrix(y_test,y_pred)
```

```python
#fitting bernoulli naive bayes to the training set
from sklearn.naive_bayes import BernoulliNB
Bclassifier=BernoulliNB()
Bclassifier.fit(x_train,y_train)
y_pred=Bclassifier.predict(x_test)

from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy is:',accuracy_score(y_test,y_pred))

confusion_matrix(y_test,y_pred)
```

```python
#fitting multinomial naive bayes to the training set
from sklearn.naive_bayes import MultinomialNB
Mclassifier=MultinomialNB()
Mclassifier.fit(x_train,y_train)
y_pred=Mclassifier.predict(x_test)

from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
print('Accuracy is:',accuracy_score(y_test,y_pred))

confusion_matrix(y_test,y_pred)
```
